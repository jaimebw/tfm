```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
%matplotlib widget
data1= ["../data/pca_results/skew_1_pca_labels.pkl","../data/pca_results/mean_abs_1_pca_labels.pkl","../data/pca_results/kurt_1_pca_labels.pkl","../data/pca_results/sfactor_1_pca_labels.pkl","../data/pca_results/rms_1_pca_labels.pkl"]
data1 = [Path(i) for i in data1]
fig, axes =plt.subplots(len(data1),1,figsize = (12,18),sharex = True,sharey= True)
for index,i in enumerate(data1):
    df = pd.read_pickle(i)
    df.plot(ax = axes[index])
    axes[index].set_title(f"{i.stem}; Tresh = {df.thresh.values[0]}")
```
