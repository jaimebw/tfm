```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import re
input_path = Path("../data/pca_results")
```

```python
data1 = []
data2 = []
data3 = []
for data in input_path.glob("*.pkl"):
    if "1" in re.findall(r"\d+",str(data)):
        data1.append(data)
    elif "2" in re.findall(r"\d+",str(data)):
        data2.append(data)
    else:
        data3.append(data)
data1.sort()
data2.sort()
data3.sort()
```

```python
data1
```

```python
%matplotlib widget
fig, axes =plt.subplots(len(data1),1,figsize = (12,18),sharex = True,sharey= True)
for index,i in enumerate(data1):
    df = pd.read_pickle(i)
    df.plot(ax = axes[index],logy = True)
    axes[index].set_title(f"{i.stem}; Tresh = {df.thresh.values[0]}")
```

```python
%matplotlib widget
fig, axes =plt.subplots(len(data2),1,figsize = (12,24),sharex = True,sharey= True)
for index,i in enumerate(data2):
    df = pd.read_pickle(i)
    df.plot(ax = axes[index],ylim = [0,5])
    axes[index].set_title(f"{i.stem}; Tresh = {df.thresh.values[0]}")
```

```python
%matplotlib widget
fig, axes =plt.subplots(len(data3),1,figsize = (12,24),sharex = True,sharey= True)
for index,i in enumerate(data3):
    df = pd.read_pickle(i)
    df.plot(ax = axes[index],ylim = [0,5])
    axes[index].set_title(f"{i.stem}; Tresh = {df.thresh.values[0]}")
```

```python
font_dirs = ['../otros/lmr']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    font_manager.fontManager.addfont(font_file)
matplotlib.style.context("seaborn-paper")
plt.rcParams['font.family'] = 'Latin Modern Roman'
myFmt = mdates.DateFormatter("%d/%m")

lol = "../data/pca_results/rolling_avg_50_1_pca_labels.pkl"
with plt.style.context("seaborn-paper"):
    fig,ax= plt.subplots(1,1,figsize = (9,5))
    df = pd.read_pickle(lol)
    ax.plot(df["dist_mob"],label = "Mahal distance")
    ax.plot(df["thresh"],label = "Anomaly threshold",color = "red",linestyle = "--")
    ax.legend(prop={'size': 12})
    ax.set_ylim([0,5])
    ax.xaxis.set_major_formatter(myFmt)
    ax.set_xlabel("Days")
    ax.annotate('Real anomaly', xy=(df.loc[df.anomaly == True].index[13], df.thresh.values[0]),  xycoords='data',
            xytext=(0.4, 0.4), textcoords='axes fraction',
            arrowprops=dict(arrowstyle="fancy",connectionstyle="arc3",color ="green"),
            horizontalalignment='right', verticalalignment='top',
            )
    ax.annotate('Noise', xy=(df.loc[df.anomaly == True].index[0], df.thresh.values[0]),  xycoords='data',
            xytext=(0.2, 0.4), textcoords='axes fraction',
            arrowprops=dict(arrowstyle="fancy",connectionstyle="arc3",color ="red"),
            horizontalalignment='right', verticalalignment='top',
            )
    fig.savefig("../figures/mahal_just_ma50_1.pdf",dpi = 300)
```
